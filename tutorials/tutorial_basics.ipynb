# Basic Usage and Manipulation with Units

We start by creating unit types representing atomic quantities such as Mass or Force. We will manipulate with these types using basic arithmetics and comparisons.

```python
from engunits import Acceleration, Mass
```

We start with definition of some masses that are either with different values or different units. Notice that $m_1$ and $m_3$, while defined differently, actually represent the same amount of mass.

```python
m1 = Mass(10) # default assignment of units in 'kg'
m2 = Mass(5, 'lb') # explicit assignment of units in 'lb'
m3 = Mass(10e3, 'g') # explicit assignment of units in 'g'
```

We can also define acceleration and an array of accelerations.

```python
import numpy as np

g = Acceleration(9.80665, "m/s**2") # standard acceleration of gravity
a_array = Acceleration(np.array([0, 1, 2]), "m/s**2") # array of accelerations
```

Let's work with these units now. First, let's show some arithmetic work, comparison and overall handy features when working with these unit types.

```python
print(f"Directly using a unit object in a print statement shows the value and units: m1 = {m1}.")
print(f"Units can be accessed as a string: m1.units = '{m1.units}'.")
print(f"The value of the mass can be accessed as a float: m1.value = {m1.value}.")
print(f"Units can be converted using the .to() method: m1.to('lb') = {m1.to('lb'):.2f}.")
print(f"Unit can be called directly in-place to convert the value: m1('lb') = {m1('lb'):.2f}.")
```

```python
if m1 == m3:
    print("Comparing 2 different variables with different units but the same resulting value shows that they are equal: m1 == m3 is True.")
```
